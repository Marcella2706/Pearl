```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("/kaggle/input/heart-failure-prediction/heart.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler


le = LabelEncoder()
scaler = StandardScaler()
num_cols = [col for col in df.select_dtypes(include=['int64', 'float64']).columns if col != "HeartDisease"]

# for col in num_cols:
#     df[col] = scaler.fit_transform(df[[col]])


for col in df.select_dtypes(include=['object', 'category']).columns:
    df[col] = le.fit_transform(df[col])

print(df)
    
```

```python

X = df.drop(['HeartDisease'], axis = 1)
y = df['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

print(model.feature_importances_)
sorted_index  = np.argsort(model.feature_importances_);

print(sorted_index)
```

```python
sorted_features = X_train.iloc[:, sorted_index]
```

```python
import optuna
from sklearn.model_selection import train_test_split, cross_val_score
import xgboost as xgb
from sklearn.ensemble import VotingClassifier



Random_forest =  {'n_estimators': 455, 'max_depth': 6, 'min_samples_split': 10, 'min_samples_leaf': 2, 'max_features': 'log2'}
logistic =  {'C': 0.0016287456348191987, 'solver': 'saga'}
XGBoost = {'booster': 'dart', 'lambda': 5.086991897983594, 'alpha': 0.1433739176157671, 'max_depth': 13, 'eta': 0.03329359933322064, 'subsample': 0.8924852317093961, 'colsample_bytree': 0.6761707122251873, 'min_child_weight': 2}
log_clf = LogisticRegression()
rf_clf = RandomForestClassifier(**Random_forest)

xg_boast = xgb.XGBClassifier(**XGBoost)

ensemble_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rf_clf), ('xg', xg_boast)],
    voting='soft'
)
```
